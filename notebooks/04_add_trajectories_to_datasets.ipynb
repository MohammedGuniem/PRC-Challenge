```python
import pandas as pd
pd.set_option('display.max_columns', None)  # Use None to display all columns
```

<h3>Read summarized_trajectories</h3>

```python
summarized_trajectory_df = pd.read_csv('./data/summarized_trajectory.csv')
display(summarized_trajectory_df)
```

<h3>Add trajectories to challenge set</h3>

```python
challenge_set_df = pd.read_csv('./data/processed_challenge_set.csv')
display(challenge_set_df)
```

```python
merged_challenge_df = pd.merge(challenge_set_df, summarized_trajectory_df, on='flight_id', how='left')
display(merged_challenge_df)
```

```python
rows_with_nan = merged_challenge_df[merged_challenge_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_challenge_df.fillna(0, inplace=True)

rows_with_nan = merged_challenge_df[merged_challenge_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_challenge_df.to_csv("./data/challenge_set_with_trajectories.csv", index=False)
print(f"./data/challenge_set_with_trajectories.csv is saved!")
```

<h3>Add trajectories to submission set</h3>

```python
submission_set_df = pd.read_csv('./data/processed_submission_set.csv')
display(submission_set_df)
```

```python
merged_submission_df = pd.merge(submission_set_df, summarized_trajectory_df, on='flight_id', how='left')

merged_submission_df.drop('tow', axis=1, inplace=True)

display(merged_submission_df)
```

```python
rows_with_nan = merged_submission_df[merged_submission_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_submission_df.fillna(0, inplace=True)

rows_with_nan = merged_submission_df[merged_submission_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_submission_df.to_csv("./data/submission_set_with_trajectories.csv", index=False)
print(f"./data/submission_set_with_trajectories.csv is saved!")
```

<h3>Add trajectories to final submission set</h3>

```python
final_submission_set_df = pd.read_csv('./data/processed_final_submission_set.csv')
display(final_submission_set_df)
```

```python
merged_final_submission_df = pd.merge(final_submission_set_df, summarized_trajectory_df, on='flight_id', how='left')

merged_final_submission_df.drop('tow', axis=1, inplace=True)

display(merged_final_submission_df)
```

```python
rows_with_nan = merged_final_submission_df[merged_final_submission_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_final_submission_df.fillna(0, inplace=True)

rows_with_nan = merged_final_submission_df[merged_final_submission_df.isna().any(axis=1)]
display(rows_with_nan)
```

```python
merged_final_submission_df.to_csv("./data/final_submission_set_with_trajectories.csv", index=False)
print(f"./data/final_submission_set_with_trajectories.csv is saved!")
```
