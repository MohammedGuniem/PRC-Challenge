```python
import pandas as pd
```

<h3>Read summarized_trajectories</h3>

```python
summarized_trajectory_df = pd.read_csv('./data/summarized_trajectory.csv')
display(summarized_trajectory_df)
```

<h3>Add trajectories to challenge set</h3>

```python
challenge_set_df = pd.read_csv('./data/processed_challenge_set.csv')
display(challenge_set_df)
```

```python
merged_challenge_df = pd.merge(challenge_set_df, summarized_trajectory_df, on='flight_id', how='left')
display(merged_challenge_df)
```

```python
merged_challenge_df.to_csv("./data/challenge_set_with_trajectories.csv", index=False)
print(f"./data/challenge_set_with_trajectories.csv is saved!")
```

<h3>Add trajectories to submission set</h3>

```python
submission_set_df = pd.read_csv('./data/processed_submission_set.csv')
display(submission_set_df)
```

```python
merged_submission_df = pd.merge(submission_set_df, summarized_trajectory_df, on='flight_id', how='left')
display(merged_submission_df)
```

```python
merged_submission_df.to_csv("./data/submission_set_with_trajectories.csv", index=False)
print(f"./data/submission_set_with_trajectories.csv is saved!")
```
