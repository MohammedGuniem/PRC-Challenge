<h1>Columns vs. Target Correlations</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
one_hot_encoded_challenge_set_df = pd.read_csv('../data/one_hot_encoded_challenge_set_df.csv')
display(one_hot_encoded_challenge_set_df)
```

```python
excluded_columns = ['tow', 'aircraft_max_weight']
correlations = []
for column in one_hot_encoded_challenge_set_df.columns:
    if column not in excluded_columns:
        corr = one_hot_encoded_challenge_set_df[[column, 'tow']].corr().iloc[0, 1]
        if corr > 0.01 or corr < -0.01:
            print(f"correlation between '{column}' and target 'tow' is {corr}")
            correlations.append(column)

print(len(correlations))
print(correlations)
```
