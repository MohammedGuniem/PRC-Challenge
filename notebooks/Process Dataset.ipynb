<h1>I - Process Dataset</h1>

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

<h3>A - View Row Dataset</h3>

```python
data_df = pd.read_csv('../data/challenge_set.csv')
print("Row DataFrame (challenge_set.csv):")
display(data_df)
```

<h3>B - Process Dataset</h3>

```python
processed_challenge_set_df = data_df.loc[:, ['flight_id', 'date', 'adep', 'country_code_adep', 'ades', 'country_code_ades', 'aircraft_type', 'wtc', 'airline', 'flight_duration', 'taxiout_time', 'flown_distance', 'tow']]

# Split date into 2 columns, year is not important since it is the same in the entire dataset 2022
processed_challenge_set_df['date'] = pd.to_datetime(processed_challenge_set_df['date'])
processed_challenge_set_df['month-day'] = processed_challenge_set_df['date'].dt.strftime('%m-%d')
processed_challenge_set_df['month'] = processed_challenge_set_df['date'].dt.strftime('%m').astype(int)
processed_challenge_set_df['day_in_month'] = processed_challenge_set_df['date'].dt.strftime('%d').astype(int)

# Drop the original date column
processed_challenge_set_df = processed_challenge_set_df.drop(columns=['date'])

print("Processed DataFrame (processed_challenge_set.csv):")
display(processed_challenge_set_df)

path = "../data/processed_challenge_set.csv"
processed_challenge_set_df.to_csv(path, index=False)
print("Saved Processed DataFrame to path:", path)
```

<h3>C - Add Maximum Aircraft Weights</h3>

```python
aircraft_max_weights = pd.read_csv("../data/aircraft_max_weights.csv")[['Type', 'MTOW [kg]']]
print("Row (aircraft_max_weights.csv) DataFrame:")
display(aircraft_max_weights)

combined_challenge_set_df = processed_challenge_set_df.copy()
combined_challenge_set_df['aircraft_max_weight'] = np.inf
for index, row in combined_challenge_set_df.iterrows():
    aircraft_type = row['aircraft_type']
    for _, aircraft_max_weights_row in aircraft_max_weights.iterrows():
        if aircraft_type in aircraft_max_weights_row['Type']:
            combined_challenge_set_df.at[index, 'aircraft_max_weight'] = np.int64(aircraft_max_weights_row['MTOW [kg]'])
            break
print("Combined (challenge_set.csv & aircraft_max_weights.csv) DataFrame:")
display(combined_challenge_set_df)

path = "../data/combined_challenge_set.csv"
combined_challenge_set_df.to_csv(path, index=False)
print("Saved Combined DataFrame (challenge_set.csv & aircraft_max_weights.csv) to path:", path)
```

<h1>II - Encode Dataset</h1>

<h3>A - Label Encoding With Value Recognition Across Columns</h3>

```python
# Identify string columns
string_columns = combined_challenge_set_df.select_dtypes(include=['object']).columns

# Create a global mapping for encoding
unique_values = pd.unique(combined_challenge_set_df[string_columns].values.ravel('K'))
value_to_int = {value: idx for idx, value in enumerate(unique_values)}

# Define a function to encode values
def encode_value(value):
    return value_to_int.get(value, value)  # Encode string values, keep other values unchanged

# Apply encoding only to string columns
label_encoded_with_value_recognition_challenge_set_df = combined_challenge_set_df.copy()  # Create a copy of the original DataFrame
label_encoded_with_value_recognition_challenge_set_df[string_columns] = label_encoded_with_value_recognition_challenge_set_df[string_columns].map(encode_value)

print("Label Encoded DataFrame With Value Recognition Across Columns:")
display(label_encoded_with_value_recognition_challenge_set_df)

# Store the result in a CSV file
path = "../data/label_encoded_with_value_recognition_challenge_set.csv"
label_encoded_with_value_recognition_challenge_set_df.to_csv(path, index=False)
print("Saved Encoded DataFrame to path:", path)
```

<h3>B - Label Encoding Without Value Recognition Across Columns</h3>

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Create a copy of the original DataFrame to hold the encoded data
label_encoded_without_value_recognition_challenge_set_df = combined_challenge_set_df.copy()

# Select only categorical columns and apply LabelEncoder
label_encoded_without_value_recognition_challenge_set_df[label_encoded_without_value_recognition_challenge_set_df.select_dtypes(include=['object']).columns] = label_encoded_without_value_recognition_challenge_set_df.select_dtypes(include=['object']).apply(lambda col: le.fit_transform(col))

# Display the new encoded DataFrame
display(label_encoded_without_value_recognition_challenge_set_df)

# Store the result in a CSV file
path = "../data/label_encoded_without_value_recognition_challenge_set.csv"
label_encoded_without_value_recognition_challenge_set_df.to_csv(path, index=False)
print("Saved Encoded DataFrame to path:", path)
```

<h3>C - One-Hot Encoding</h3>

```python
# Identify categorical columns
categorical_columns = ['adep', 'country_code_adep', 'ades', 'country_code_ades', 'aircraft_type', 'wtc', 'airline', 'month-day']

# Apply one-hot encoding to the categorical columns
one_hot_encoded_df = pd.get_dummies(combined_challenge_set_df, columns=categorical_columns)

print("One-Hot Encoded DataFrame:")
display(one_hot_encoded_df)
    
# Store the result in a CSV file
path = "../data/one_hot_encoded_challenge_set.csv"
one_hot_encoded_df.to_csv(path, index=False)
print("Saved Encoded DataFrame to path:", path)
```

<h3>D - Target Encoding</h3>

```python
# Identify categorical columns
categorical_cols = combined_challenge_set_df.select_dtypes(include=['object', 'category']).columns

# Create a copy of the original DataFrame
target_encoded_df = combined_challenge_set_df.copy()  

# Apply target encoding
for col in categorical_cols:
    # Compute the mean of the target variable for each category
    means = target_encoded_df.groupby(col)['tow'].mean()
    # Map the means to the original categorical column
    target_encoded_df[col + '_encoded'] = target_encoded_df[col].map(means)

# Drop the original categorical columns if desired
target_encoded_df = target_encoded_df.drop(columns=categorical_cols)

print("Target Encoded DataFrame:")
display(target_encoded_df)

# Store the result in a CSV file
path = "../data/target_encoded_challenge_set.csv"
target_encoded_df.to_csv(path, index=False)
print("Saved Encoded DataFrame to path:", path)
```
