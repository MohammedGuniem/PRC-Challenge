<h3>Encoding</h3>

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
challenge_set = pd.read_csv('../data/challenge_set.csv')

# Split date into 2 columns, year is not important since it is the same in the entire dataset 2022
challenge_set['date'] = pd.to_datetime(challenge_set['date'])
challenge_set['month-day'] = challenge_set['date'].dt.strftime('%m-%d')
challenge_set['month'] = challenge_set['date'].dt.strftime('%m').astype(int)
challenge_set['day_in_month'] = challenge_set['date'].dt.strftime('%d').astype(int)

challenge_features = challenge_set[['flight_id', 'adep', 'country_code_adep', 'ades', 'country_code_ades', 'aircraft_type', 'wtc', 'airline', 'flight_duration', 
                          'taxiout_time', 'flown_distance', 'month-day', 'month', 'day_in_month', 'tow']]
display(challenge_features)

submission_set = pd.read_csv('../data/submission_set.csv')

# Split date into 2 columns, year is not important since it is the same in the entire dataset 2022
submission_set['date'] = pd.to_datetime(submission_set['date'])
submission_set['month-day'] = submission_set['date'].dt.strftime('%m-%d')
submission_set['month'] = submission_set['date'].dt.strftime('%m').astype(int)
submission_set['day_in_month'] = submission_set['date'].dt.strftime('%d').astype(int)

submission_features = submission_set[['flight_id', 'adep', 'country_code_adep', 'ades', 'country_code_ades', 'aircraft_type', 'wtc', 'airline', 'flight_duration', 
                          'taxiout_time', 'flown_distance', 'month-day', 'month', 'day_in_month', 'tow']]
display(submission_features)
```

```python
# Append df2 to df1
df_combined = pd.concat([challenge_features, submission_features], ignore_index=True)

# Reset the index
df_combined.reset_index(drop=True, inplace=True)

display(df_combined)
```

```python
#unique_aircraft_type_values = df_combined['aircraft_type'].unique()
#print(unique_aircraft_type_values)

# Calculate min and max takeoff weights for each aircraft type
df_combined['min_tow'] = df_combined.groupby('aircraft_type')['tow'].transform('min')
df_combined['max_tow'] = df_combined.groupby('aircraft_type')['tow'].transform('max')

display(df_combined)
```

```python
def encode(df: pd.DataFrame, categorical_columns: list[str]) -> pd.DataFrame:
    encoded_df = df.copy()
    label_encoder = LabelEncoder()

    for column in encoded_df.columns:
        if column in categorical_columns:     
            encoded_df[column] = label_encoder.fit_transform(df[column])
    
    return encoded_df

categorical_columns = [
    "adep",
    "country_code_adep",
    "ades",
    "country_code_ades",
    "aircraft_type",
    "wtc",
    "airline",
    "month-day"
]

encoded_df = encode(df=df_combined, categorical_columns=categorical_columns)

display(encoded_df)
```

```python
def separate_data(original_df: pd.DataFrame, index: int):
    encoded_challenge_set = original_df.iloc[:index].reset_index(drop=True)
    encoded_submission_set = original_df.iloc[index:].reset_index(drop=True)
    return encoded_challenge_set, encoded_submission_set

encoded_challenge_set, encoded_submission_set = separate_data(original_df=encoded_df, index=len(challenge_features))

display(encoded_challenge_set)

display(encoded_submission_set)
```

```python
def save_dataframe(path: str, df: pd.DataFrame) -> None:
    df.to_csv(path, index=False)
    print(f"{path} is saved!")

save_dataframe(path="../data/encoded_challenge_set.csv", df=encoded_challenge_set)
save_dataframe(path="../data/encoded_submission_set.csv", df=encoded_submission_set)
```
