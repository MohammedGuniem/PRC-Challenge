```python
from IPython.display import display
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_airliners_by_maximum_takeoff_weight')
df = data[0]
display(df)
```

```python
# Save the correct table as a CSV file
csv_path = "data/aircraft_max_weights.csv"
df.to_csv(csv_path, index=False)

# Load the CSV back into a DataFrame and display it
df = pd.read_csv("data/aircraft_max_weights.csv")
display(df)
```

```python
data_df = pd.read_csv('data/challenge_set.csv')

challenge_set_df = data_df.loc[:, ['adep', 'country_code_adep', 'ades', 'country_code_ades', 'aircraft_type', 'wtc', 'airline', 'flight_duration', 'taxiout_time', 'flown_distance', 'tow']]

aircraft_types = challenge_set_df['aircraft_type'].unique()
print(aircraft_types)

value_counts = challenge_set_df['aircraft_type'].value_counts()
print(value_counts)

print(value_counts['A320'])
```

```python
matching_rows, non_matching_rows = 0, 0
for aircraft_type in aircraft_types:
    print("aircraft type: ", aircraft_type)
    filtered = df[df['Type'].str.contains(aircraft_type, case=False, na=False)]
    if len(filtered) > 0:
        matching_rows += value_counts[aircraft_type]
        display(filtered)
    else:
        non_matching_rows += value_counts[aircraft_type]

print(f"{len(challenge_set_df)=}")
print(f"{matching_rows=}")
print(f"{non_matching_rows=}")
print(f"{(matching_rows+non_matching_rows)=}")
print(f"{(matching_rows/len(challenge_set_df))=}")
```
