```python
trajactory_df = pq.ParquetDataset('data/2022-01-01.parquet').read().to_pandas()

display(trajactory_df.head())
```

```python
print("Number of rows in trajactory_df:", len(trajactory_df))

unique_count = trajactory_df['flight_id'].nunique()
print("Number of unique flight_id values:", unique_count)

unique_count = trajactory_df['icao24'].nunique()
print("Number of unique icao24 values:", unique_count)

filtered_df = trajactory_df[trajactory_df['flight_id'].apply(lambda x: isinstance(x, object))]
print("Number of cell values with object type for flight_id:", len(filtered_df))

trajactory_df['flight_id'] = trajactory_df['flight_id'].astype('int64')

filtered_df = trajactory_df[trajactory_df['flight_id'].apply(lambda x: isinstance(x, int))]
print("Number of cell values with int type for flight_id:", len(filtered_df))

result = trajactory_df[trajactory_df['flight_id'] == 248763780]
display(result)

result = trajactory_df[trajactory_df['flight_id'] == 248753821]
display(result)
```
