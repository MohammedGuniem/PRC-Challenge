```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import json

pd.set_option('display.max_columns', None)  # Use None to display all columns
```

<h3>Append processed submission set to processed challenge set</h3>

```python
challenge_set_df = pd.read_csv('./data/challenge_set_with_trajectories.csv')
display(challenge_set_df)
```

```python
submission_set_df = pd.read_csv('./data/submission_set_with_trajectories.csv')
display(submission_set_df)
```

```python
final_submission_set_df = pd.read_csv('./data/final_submission_set_with_trajectories.csv')
display(final_submission_set_df)
```

```python
df_combined = pd.concat([challenge_set_df, submission_set_df, final_submission_set_df], ignore_index=True)
display(df_combined)
```

<h3>label encode them</h3>

```python
def encode(df: pd.DataFrame, categorical_columns: list[str]) -> pd.DataFrame:
    encoded_df = df.copy()
    label_encoder = LabelEncoder()
    for column in encoded_df.columns:
        if column in categorical_columns:     
            encoded_df[column] = label_encoder.fit_transform(df[column])
    return encoded_df
```

```python
categorical_columns = [
    'month_day',
    'actual_offblock_time_hour_minute',
    'adep',
    'country_code_adep',
    'arrival_time_hour_minute',
    'ades', 
    'country_code_ades', 
    'aircraft_type', 
    'wtc', 
    'airline', 
]
encoded_df = encode(df=df_combined, categorical_columns=categorical_columns)
display(encoded_df)
```

<h3>Split them back into encoded_challenge_set.csv and encoded_submission_set.csv</h3>

```python
def separate_data(original_df: pd.DataFrame, challenge_set_len: int, submission_set_len: int):
    encoded_challenge_set = original_df.iloc[:challenge_set_len].reset_index(drop=True)
    encoded_submission_set = original_df.iloc[challenge_set_len:(challenge_set_len+submission_set_len)].reset_index(drop=True)
    encoded_final_submission_set = original_df.iloc[(challenge_set_len+submission_set_len):].reset_index(drop=True)
    return encoded_challenge_set, encoded_submission_set, encoded_final_submission_set
```

```python
encoded_challenge_set, encoded_submission_set, encoded_final_submission_set = separate_data(
    original_df=encoded_df, 
    challenge_set_len=len(challenge_set_df), 
    submission_set_len=len(submission_set_df)
)
```
