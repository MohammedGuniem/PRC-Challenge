```python
submission_set = pd.read_csv('../../data/submission_set_with_trajectory.csv')

# Split date into 2 columns, year is not important since it is the same in the entire dataset 2022
submission_set['date'] = pd.to_datetime(submission_set['date'])
submission_set['month-day'] = submission_set['date'].dt.strftime('%m-%d')
submission_set['month'] = submission_set['date'].dt.strftime('%m').astype(int)
submission_set['day_in_month'] = submission_set['date'].dt.strftime('%d').astype(int)

# Convert the column to datetime type
submission_set['actual_offblock_time'] = pd.to_datetime(submission_set['actual_offblock_time'])
# Extract hour, minute, and create combined hour:minute column
submission_set['actual_offblock_time_hour'] = submission_set['actual_offblock_time'].dt.hour
submission_set['actual_offblock_time_minute'] = submission_set['actual_offblock_time'].dt.minute
submission_set['actual_offblock_time_hour_minute'] = submission_set['actual_offblock_time'].dt.strftime('%H:%M')

# Convert the column to datetime type
submission_set['arrival_time'] = pd.to_datetime(submission_set['arrival_time'])
# Extract hour, minute, and create combined hour:minute column
submission_set['arrival_time_hour'] = submission_set['arrival_time'].dt.hour
submission_set['arrival_time_minute'] = submission_set['arrival_time'].dt.minute
submission_set['arrival_time_hour_minute'] = submission_set['arrival_time'].dt.strftime('%H:%M')

submission_features = submission_set[columns]

display(submission_features)
```

```python
# Append df2 to df1
df_combined = pd.concat([challenge_features, submission_features], ignore_index=True)

# Reset the index
df_combined.reset_index(drop=True, inplace=True)

display(df_combined)
```

```python
def encode(df: pd.DataFrame, categorical_columns: list[str]) -> pd.DataFrame:
    encoded_df = df.copy()
    label_encoder = LabelEncoder()

    for column in encoded_df.columns:
        if column in categorical_columns:     
            encoded_df[column] = label_encoder.fit_transform(df[column])
            
            if column == 'aircraft_type':
                aircraft_types_encoding = {encoded: original for original, encoded in zip(label_encoder.classes_, range(len(label_encoder.classes_)))}
                with open('../../data/aircraft_types_encoding.json', 'w') as json_file:
                    json.dump(aircraft_types_encoding, json_file, indent=4)

    
    return encoded_df

categorical_columns = [
    'month-day',
    'adep',
    'country_code_adep', 
    'ades', 
    'country_code_ades', 
    'aircraft_type', 
    'wtc', 
    'airline', 
    'actual_offblock_time_hour_minute', 
    'arrival_time_hour_minute'
]

encoded_df = encode(df=df_combined, categorical_columns=categorical_columns)

display(encoded_df)
```

```python
def separate_data(original_df: pd.DataFrame, index: int):
    encoded_challenge_set = original_df.iloc[:index].reset_index(drop=True)
    encoded_submission_set = original_df.iloc[index:].reset_index(drop=True)
    return encoded_challenge_set, encoded_submission_set

encoded_challenge_set, encoded_submission_set = separate_data(original_df=encoded_df, index=len(challenge_features))

display(encoded_challenge_set)

display(encoded_submission_set)
```

```python
def save_dataframe(path: str, df: pd.DataFrame) -> None:
    df.to_csv(path, index=False)
    print(f"{path} is saved!")
```
